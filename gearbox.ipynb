```python
"""
選択的マハラノビス AE モードを組み込んだフルスクリプト
---------------------------------------------------------------
1. AudioDataset     : (tensor, domain)  /  (tensor, label, domain)
2. 逆共分散行列推定 : source / target 正常データで μ・Σ⁻¹ を計算
3. 評価             : min{D_s, D_t} を異常スコアに採用
"""

import os, random, time, warnings
import librosa
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset

warnings.filterwarnings("ignore")

# --------------------------- 1  ユーティリティ --------------------------- #
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False


# --------------------------- 2  データセット --------------------------- #
class AudioDataset(Dataset):
    """
    file name 例:
        gearbox_source_0001.wav
        gearbox_target_0002_anomaly.wav
    """
    def __init__(self, folder_path, is_train=True, feature_type="mfcc"):
        self.folder_path   = folder_path
        self.file_paths    = sorted([os.path.join(folder_path, f)
                                     for f in os.listdir(folder_path)
                                     if f.endswith(".wav")])
        self.is_train      = is_train
        self.feature_type  = feature_type

        self.domains = ["source" if "_source_" in os.path.basename(p)
                        else "target" for p in self.file_paths]

        if not is_train:   # テスト時のみラベルを保持
            self.labels = [1 if "anomaly" in os.path.basename(p) else 0
                           for p in self.file_paths]

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        path = self.file_paths[idx]
        y, sr = librosa.load(path, sr=SAMPLE_RATE)

        if self.feature_type == "mfcc":
            mfcc = librosa.feature.mfcc(
                y=y, sr=sr, n_mfcc=N_MFCC, n_fft=N_FFT, hop_length=HOP_LENGTH
            )
            feature = (mfcc - mfcc.mean()) / (mfcc.std() + 1e-9)
            # 横幅を 8 の倍数へ切り詰め
            T = feature.shape[1]
            feature = feature[:, : T - (T % 8)]
        else:
            raise ValueError("Only 'mfcc' supported")

        feature_tensor = torch.tensor(feature, dtype=torch.float32).unsqueeze(0)

        domain = self.domains[idx]
        if self.is_train:
            return feature_tensor, domain
        else:
            label = self.labels[idx]
            return feature_tensor, label, domain


# --------------------------- 3  Autoencoder --------------------------- #
class Autoencoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1), nn.BatchNorm2d(32), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.bottleneck = nn.Sequential(
            nn.Conv2d(128, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(256, 128, 2, stride=2),
            nn.BatchNorm2d(128), nn.ReLU(),
            nn.ConvTranspose2d(128, 64, 2, stride=2),
            nn.BatchNorm2d(64), nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 2, stride=2),
            nn.BatchNorm2d(32), nn.ReLU(),
            nn.Conv2d(32, 1, 3, padding=1)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.bottleneck(x)
        x = self.decoder(x)
        return x


# --------------------------- 4  学習 --------------------------- #
def train_autoencoder(model, train_loader, val_loader):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    train_losses, val_losses = [], []
    pbar = tqdm(range(EPOCHS), desc="Train", unit="epoch")

    for _ in pbar:
        model.train()
        tr_loss = 0.0
        for batch, _ in train_loader:
            batch = batch.to(DEVICE)
            recon = model(batch)
            loss  = criterion(recon, batch)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            tr_loss += loss.item()

        model.eval()
        vl_loss = 0.0
        with torch.no_grad():
            for vbatch, _ in val_loader:
                vbatch = vbatch.to(DEVICE)
                vl_loss += criterion(model(vbatch), vbatch).item()

        train_losses.append(tr_loss / len(train_loader))
        val_losses.append(vl_loss / len(val_loader))
        pbar.set_postfix(Train=train_losses[-1], Val=val_losses[-1])

    return train_losses, val_losses


# --------------------------- 5  マハラノビス --------------------------- #
def compute_mahalanobis_stats(model, loader):
    model.eval()
    errors = []
    with torch.no_grad():
        for batch, _ in loader:
            batch = batch.to(DEVICE)
            diff  = (batch - model(batch)).view(batch.size(0), -1)
            errors.append(diff.cpu().numpy())
    errors = np.concatenate(errors, axis=0)
    mu   = errors.mean(0, keepdims=True)
    cov  = np.cov(errors.T) + 1e-6 * np.eye(errors.shape[1])
    inv  = np.linalg.inv(cov)
    return mu, inv

def mahalanobis(v, mu, inv_cov):
    diff = v - mu
    return float(diff @ inv_cov @ diff.T)


# --------------------------- 6  評価 --------------------------- #
def evaluate_mahalanobis(model, loader, mu_s, inv_s, mu_t, inv_t):
    model.eval()
    scores, labels = [], []

    with torch.no_grad():
        for batch, label, domain in tqdm(loader, desc="Eval"):
            batch = batch.to(DEVICE)
            recon = model(batch)
            diff  = (batch - recon).view(batch.size(0), -1).cpu().numpy()

            for vec, dom in zip(diff, domain):
                Ds = mahalanobis(vec, mu_s, inv_s)
                Dt = mahalanobis(vec, mu_t, inv_t)
                scores.append(min(Ds, Dt))
            labels.extend(label.numpy())

    auc = roc_auc_score(labels, scores)
    print(f"Mahalanobis‑AUC = {auc:.4f}")

    # ヒストグラム可視化
    scores = np.array(scores); labels = np.array(labels)
    plt.figure(figsize=(8,4))
    plt.hist(scores[labels==0], bins=50, alpha=.6, label="Normal", density=True)
    plt.hist(scores[labels==1], bins=50, alpha=.6, label="Anomaly", density=True)
    plt.title("Mahalanobis Distance Histogram"); plt.xlabel("Distance"); plt.ylabel("Density")
    plt.legend(); plt.grid(); plt.tight_layout(); plt.show()


def plot_losses(tr, vl):
    plt.figure(figsize=(8,4))
    plt.plot(tr, label="Train"); plt.plot(vl, label="Val")
    plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("MSE Loss"); plt.legend()
    plt.grid(); plt.tight_layout(); plt.show()


# --------------------------- 7  ハイパーパラメータ --------------------------- #
SAMPLE_RATE = 16000
N_MFCC      = 16
HOP_LENGTH  = 512
N_FFT       = 1024

BATCH_SIZE  = 32
EPOCHS      = 50
DEVICE      = torch.device("mps" if torch.backends.mps.is_available()
                           else "cuda" if torch.cuda.is_available()
                           else "cpu")

# --------------------------- 8  メイン --------------------------- #
if __name__ == "__main__":
    print("device:", DEVICE)
    set_seed()

    # ---------- 8‑1  データ用意 ---------- #
    train_folders = ["dev/gearbox/train"]
    datasets = [AudioDataset(f, is_train=True) for f in train_folders]
    full_dataset = ConcatDataset(datasets)

    tr_size = int(0.8 * len(full_dataset))
    va_size = len(full_dataset) - tr_size
    train_dataset, val_dataset = random_split(full_dataset, [tr_size, va_size])

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE, shuffle=False)

    test_dataset = AudioDataset("dev/gearbox/test", is_train=False)
    test_loader  = DataLoader(test_dataset, batch_size=1, shuffle=False)

    # ---------- 8‑2  学習 ---------- #
    model = Autoencoder().to(DEVICE)
    tr_losses, va_losses = train_autoencoder(model, train_loader, val_loader)
    plot_losses(tr_losses, va_losses)

    # ---------- 8‑3  μ・Σ⁻¹ 推定 ---------- #
    # domain 別サブセット作成
    def domain_indices(dataset, dom):
        return [i for i, (_, d) in enumerate(dataset) if d == dom]

    source_loader = DataLoader(
        torch.utils.data.Subset(train_dataset,
                                domain_indices(train_dataset, "source")),
        batch_size=BATCH_SIZE, shuffle=False)
    target_loader = DataLoader(
        torch.utils.data.Subset(train_dataset,
                                domain_indices(train_dataset, "target")),
        batch_size=BATCH_SIZE, shuffle=False)

    mu_s, inv_s = compute_mahalanobis_stats(model, source_loader)
    mu_t, inv_t = compute_mahalanobis_stats(model, target_loader)

    # ---------- 8‑4  評価 ---------- #
    evaluate_mahalanobis(model, test_loader, mu_s, inv_s, mu_t, inv_t)
```

```python
"""
DCASE‑style First‑Shot Autoencoder
=================================
PDF (Harada et al., EUSIPCO 2023) の **“5 frame × 128 Log‑Mel
= 640 dim 全結合 Autoencoder”** と選択的マハラノビス距離を
そのまま実装したフルスクリプトです。

・前処理  : STFT 64 ms / 50 %Hop → 128 Log‑Mel → 5 フレーム連結
・モデル  : Linear‑AE (640→128→128→128→8→…→640)
・スコア  :  min{D_s , D_t}（Mahalanobis）
・AUC     : ファイル単位で算出（各ファイルでセグメント平均）
"""

# ------------------------------------------------------------------ #
# 1. 依存ライブラリ
# ------------------------------------------------------------------ #
import os, random, time, warnings
import numpy as np
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset

warnings.filterwarnings("ignore")

# ------------------------------------------------------------------ #
# 2. ハイパーパラメータ
# ------------------------------------------------------------------ #
SR          = 16_000          # サンプリング周波数
FRAME_LEN   = 1024            # 64 ms
HOP_LEN     = 512             # 50 % hop
N_MELS      = 128
N_FRAME     = 5               # 連結するフレーム数
SEG_DIM     = N_MELS * N_FRAME

BATCH_SIZE  = 256
EPOCHS      = 100
DEVICE      = torch.device("cuda" if torch.cuda.is_available() else
                            "mps"  if torch.backends.mps.is_available()
                                    else "cpu")

# ------------------------------------------------------------------ #
# 3. 乱数固定
# ------------------------------------------------------------------ #
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

# ------------------------------------------------------------------ #
# 4. データセット
# ------------------------------------------------------------------ #
class LogMel5FrameDataset(Dataset):
    """
    * train  : (vector, domain)        … domain ∈ {"source","target"}
    * test   : (vector, label, domain, file_id)
    ※ ファイル名に  _source_ / _target_ / _anomaly_ を含む想定
    """
    def __init__(self, folder, is_train=True):
        self.is_train = is_train
        self.segments = []          # list[dict]  (vector, domain, label, file_id)

        file_paths = sorted([os.path.join(folder, f) for f in os.listdir(folder)
                             if f.endswith(".wav")])
        for fid, path in enumerate(file_paths):
            y, _ = librosa.load(path, sr=SR)
            mel = librosa.feature.melspectrogram(
                y=y, sr=SR, n_fft=FRAME_LEN, hop_length=HOP_LEN,
                n_mels=N_MELS, power=2.0
            )
            mel = np.log10(mel + 1e-12)          # Log‑Mel
            # 標準化 (各ファイルでゼロ均し)
            mel = (mel - mel.mean()) / (mel.std() + 1e-9)

            # 5フレームごとに切り出し（オーバーラップなし）
            T = mel.shape[1] - (mel.shape[1] % N_FRAME)
            mel = mel[:, :T].reshape(N_MELS, -1, N_FRAME)  # (128, num_seg, 5)
            mel = mel.transpose(1, 0, 2).reshape(-1, SEG_DIM)  # (num_seg, 640)

            domain = "source" if "_source_" in path else "target"
            label  = 1 if (not is_train and "_anomaly" in path) else 0

            for vec in mel:
                entry = dict(vec=vec.astype(np.float32),
                             domain=domain,
                             label=label,
                             fid=fid)
                self.segments.append(entry)

    def __len__(self):
        return len(self.segments)

    def __getitem__(self, idx):
        s = self.segments[idx]
        v = torch.from_numpy(s["vec"])
        if self.is_train:
            return v, s["domain"]
        else:
            return v, s["label"], s["domain"], s["fid"]

# ------------------------------------------------------------------ #
# 5. 全結合オートエンコーダ
# ------------------------------------------------------------------ #
class LinearAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(SEG_DIM, 128), nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, 128),     nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, 128),     nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128,   8)                       # bottleneck
        )
        self.decoder = nn.Sequential(
            nn.Linear( 8, 128),       nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, 128),      nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, 128),      nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, SEG_DIM)                          # output
        )

    def forward(self, x):            # x: (B, 640)
        z = self.encoder(x)
        return self.decoder(z)

# ------------------------------------------------------------------ #
# 6. 学習
# ------------------------------------------------------------------ #
def train(model, train_loader, val_loader):
    crit = nn.MSELoss()
    opt  = optim.Adam(model.parameters(), lr=1e-3)
    tr_hist, vl_hist = [], []

    for epoch in range(1, EPOCHS+1):
        model.train()
        tr_loss = 0.0
        for vec, _ in train_loader:
            vec = vec.to(DEVICE)
            loss = crit(model(vec), vec)
            opt.zero_grad(); loss.backward(); opt.step()
            tr_loss += loss.item()
        tr_loss /= len(train_loader)

        model.eval()
        vl_loss = 0.0
        with torch.no_grad():
            for vec, _ in val_loader:
                vec = vec.to(DEVICE)
                vl_loss += crit(model(vec), vec).item()
        vl_loss /= len(val_loader)

        tr_hist.append(tr_loss); vl_hist.append(vl_loss)
        print(f"Epoch {epoch:3d}/{EPOCHS}  Train {tr_loss:.4f}  Val {vl_loss:.4f}")
    return tr_hist, vl_hist

# ------------------------------------------------------------------ #
# 7. マハラノビス
# ------------------------------------------------------------------ #
def stats_by_domain(model, loader):
    model.eval(); errs = []
    with torch.no_grad():
        for vec, _ in loader:
            vec = vec.to(DEVICE)
            diff = (vec - model(vec)).cpu().numpy()
            errs.append(diff)
    errs = np.concatenate(errs, 0)                 # (N, 640)
    mu  = errs.mean(0, keepdims=True)
    cov = np.cov(errs.T) + 1e-6*np.eye(SEG_DIM)
    inv = np.linalg.inv(cov)
    return mu, inv

def maha(vec, mu, inv):           # vec: (640,)
    d = vec - mu
    return float(d @ inv @ d.T)

# ------------------------------------------------------------------ #
# 8. 評価
# ------------------------------------------------------------------ #
def evaluate(model, loader, mu_s, inv_s, mu_t, inv_t):
    model.eval()
    seg_scores, seg_labels, seg_fids = [], [], []

    with torch.no_grad():
        for vec, label, domain, fid in tqdm(loader, desc="Eval"):
            vec   = vec.to(DEVICE)
            diff  = (vec - model(vec)).cpu().numpy()       # (B,640)
            for v, f in zip(diff, fid.numpy()):
                Ds = maha(v, mu_s, inv_s)
                Dt = maha(v, mu_t, inv_t)
                seg_scores.append(min(Ds, Dt))
            seg_labels.extend(label.numpy())
            seg_fids.extend(fid.numpy())

    # ------ ファイル単位に平均 -------
    file_scores = {}
    file_labels = {}
    for sc, lb, fid in zip(seg_scores, seg_labels, seg_fids):
        file_scores.setdefault(fid, []).append(sc)
        file_labels[fid] = lb
    scores = np.array([np.mean(v) for v in file_scores.values()])
    labels = np.array([file_labels[k] for k in sorted(file_scores)])

    auc = roc_auc_score(labels, scores)
    print(f"\nFile‑level AUC = {auc:.4f}")

    # ヒストグラム
    plt.figure(figsize=(7,4))
    plt.hist(scores[labels==0], 50, alpha=.6, label="Normal", density=True)
    plt.hist(scores[labels==1], 50, alpha=.6, label="Anomaly",density=True)
    plt.title("Mahalanobis‑Distance Histogram"); plt.xlabel("Distance")
    plt.legend(); plt.grid(); plt.tight_layout(); plt.show()

# ------------------------------------------------------------------ #
# 9. メイン
# ------------------------------------------------------------------ #
if __name__ == "__main__":
    print("device =", DEVICE)
    set_seed()

    # --- データ読込（必要に応じフォルダを変更） ---
    train_folders = ["dev/gearbox/train"]       # 例
    train_dss = [LogMel5FrameDataset(f, True) for f in train_folders]
    train_full = ConcatDataset(train_dss)

    tr_size = int(0.8*len(train_full))
    va_size = len(train_full) - tr_size
    train_ds, val_ds = random_split(train_full, [tr_size, va_size])

    train_loader = DataLoader(train_ds, BATCH_SIZE, shuffle=True, drop_last=True)
    val_loader   = DataLoader(val_ds, BATCH_SIZE, shuffle=False)

    test_ds = LogMel5FrameDataset("dev/gearbox/test", is_train=False)
    test_loader = DataLoader(test_ds, BATCH_SIZE, shuffle=False)

    # --- モデル学習 ---
    ae = LinearAE().to(DEVICE)
    train_hist, val_hist = train(ae, train_loader, val_loader)

    # --- ドメイン別統計 ---
    def domain_subset(dataset, dom):
        idx = [i for i, (_, d) in enumerate(dataset) if d == dom]
        return torch.utils.data.Subset(dataset, idx)

    src_loader = DataLoader(domain_subset(train_ds, "source"), BATCH_SIZE, shuffle=False)
    tgt_loader = DataLoader(domain_subset(train_ds, "target"), BATCH_SIZE, shuffle=False)

    mu_s, inv_s = stats_by_domain(ae, src_loader)
    mu_t, inv_t = stats_by_domain(ae, tgt_loader)

    # --- 評価 ---
    evaluate(ae, test_loader, mu_s, inv_s, mu_t, inv_t)
```

```python
"""
CNN‑AE (1‑D) + Mahalanobis  First‑Shot baseline
------------------------------------------------
* 128 Mel × 5 frame パッチを **(B, C=128, L=5)** として
  Conv1d で時間方向を畳み込み → ボトルネック 8 dim
* 再構成誤差（128×5）を Mahalanobis 距離 (source / target) で評価
"""

# ---------------- 1. 依存ライブラリ ---------------- #
import os, random, warnings
warnings.filterwarnings("ignore")

import numpy as np
import librosa, matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import roc_auc_score

import torch, torch.nn as nn, torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset

# ---------------- 2. ハイパーパラメータ ---------------- #
SR, FRAME_LEN, HOP_LEN = 16_000, 1024, 512
N_MELS, N_FRAME        = 128, 5
SEG_DIM                = N_MELS * N_FRAME     # 640
BATCH_SIZE, EPOCHS     = 256, 100
DEVICE = torch.device("cuda" if torch.cuda.is_available()
                      else "mps" if torch.backends.mps.is_available()
                      else "cpu")

# ---------------- 3. 共通関数 ---------------- #
def set_seed(seed=42):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

# ---------------- 4. データセット ---------------- #
class MelPatchDataset(Dataset):
    """戻り値 (B, C=128, L=5)"""
    def __init__(self, folder, is_train=True):
        self.is_train, self.items = is_train, []
        for fid, wav in enumerate(sorted(p for p in os.listdir(folder) if p.endswith(".wav"))):
            path = os.path.join(folder, wav)
            y,_ = librosa.load(path, sr=SR)
            mel = np.log10(
                librosa.feature.melspectrogram(
                    y=y, sr=SR, n_fft=FRAME_LEN, hop_length=HOP_LEN,
                    n_mels=N_MELS, power=2.0) + 1e-12)
            
            mel = (mel - mel.mean()) / (mel.std()+1e-9)      # (128, T)

            T = mel.shape[1] - mel.shape[1] % N_FRAME
            blocks = mel[:, :T].reshape(N_MELS, -1, N_FRAME) # (128, seg, 5)
            blocks = blocks.transpose(1,0,2)                 # (seg,128,5)

            dom  = "source" if "_source_" in wav else "target"
            lab  = 1 if (not is_train and "_anomaly" in wav) else 0
            for b in blocks.astype(np.float32):
                self.items.append((b, dom, lab, fid))

    def __len__(self): return len(self.items)

    def __getitem__(self, idx):
        b, dom, lab, fid = self.items[idx]
        x = torch.from_numpy(b)           # (128,5)
        return (x, dom) if self.is_train else (x, lab, dom, fid)

# ---------------- 5. 1‑D CNN Autoencoder ---------------- #
class CNN1dAE(nn.Module):
    def __init__(self):
        super().__init__()
        # Encoder: Conv over time (L=5)
        self.enc = nn.Sequential(
            nn.Conv1d(128, 256, 3, padding=1),
            nn.BatchNorm1d(256), nn.ReLU(),
            nn.Conv1d(256, 128, 3, padding=1),
            nn.BatchNorm1d(128), nn.ReLU())
        self.fc_enc = nn.Linear(128*5, 8)

        # Decoder: FC → reshape → ConvTranspose1d
        self.fc_dec = nn.Linear(8, 128*5)
        self.dec = nn.Sequential(
            nn.ConvTranspose1d(128, 256, 3, padding=1),
            nn.BatchNorm1d(256), nn.ReLU(),
            nn.ConvTranspose1d(256, 128, 3, padding=1))

    def forward(self, x):               # x: (B,128,5)
        z = self.enc(x).view(x.size(0), -1)
        z = self.fc_enc(z)
        y = self.fc_dec(z).view(x.size(0), 128, 5)
        y = self.dec(y)
        return y                        # same shape as input

# ---------------- 6. 学習ループ ---------------- #
def train(model, tr_loader, val_loader):
    crit, opt = nn.MSELoss(), optim.Adam(model.parameters(), lr=1e-3)
    for ep in range(1, EPOCHS+1):
        model.train(); tr=0
        for xb,_ in tr_loader:
            xb = xb.to(DEVICE)
            loss = crit(model(xb), xb)
            opt.zero_grad(); loss.backward(); opt.step(); tr += loss.item()
        model.eval(); vl=0
        with torch.no_grad():
            for xb,_ in val_loader:
                xb = xb.to(DEVICE); vl += crit(model(xb), xb).item()
        print(f"Ep{ep:3d} Train{tr/len(tr_loader):.4f}  Val{vl/len(val_loader):.4f}")

# ---------------- 7. Mahalanobis ---------------- #
def get_stats(model, loader):
    model.eval(); errs=[]
    with torch.no_grad():
        for xb,_ in loader:
            xb = xb.to(DEVICE)
            diff = (xb - model(xb)).view(xb.size(0), -1).cpu().numpy()
            errs.append(diff)
    e = np.concatenate(errs,0)
    mu = e.mean(0, keepdims=True)
    cov = np.cov(e.T)+1e-6*np.eye(SEG_DIM)
    return mu, np.linalg.inv(cov)

def maha(v, mu, inv): d=v-mu; return float(d@inv@d.T)

# ---------------- 8. 評価 ---------------- #
def evaluate(model, loader, mu_s, inv_s, mu_t, inv_t):
    model.eval(); seg_sc, seg_lb, seg_fid = [],[],[]
    with torch.no_grad():
        for xb, lb, _, fid in tqdm(loader):
            xb = xb.to(DEVICE)
            diff = (xb - model(xb)).view(xb.size(0),-1).cpu().numpy()
            for d,f in zip(diff, fid.numpy()):
                seg_sc.append(min(maha(d,mu_s,inv_s), maha(d,mu_t,inv_t)))
            seg_lb.extend(lb.numpy()); seg_fid.extend(fid.numpy())
    files={},{}
    file_sc, file_lb = {}, {}
    for s,l,f in zip(seg_sc,seg_lb,seg_fid):
        file_sc.setdefault(f,[]).append(s); file_lb[f]=l
    scores=np.array([np.mean(v) for v in file_sc.values()])
    labels=np.array([file_lb[k] for k in sorted(file_sc)])
    print("File‑level AUC =", roc_auc_score(labels,scores))

# ---------------- 9. メイン ---------------- #
if __name__=="__main__":
    print("DEVICE:", DEVICE); set_seed()

    tr_dirs = ["dev/gearbox/train"]
    tr_dss  = [MelPatchDataset(d, True) for d in tr_dirs]
    full_ds = ConcatDataset(tr_dss)

    tr_len = int(0.8*len(full_ds)); va_len = len(full_ds)-tr_len
    train_ds, val_ds = random_split(full_ds,[tr_len,va_len])

    train_dl = DataLoader(train_ds,BATCH_SIZE,shuffle=True,drop_last=True)
    val_dl   = DataLoader(val_ds,BATCH_SIZE,shuffle=False)

    test_ds  = MelPatchDataset("dev/gearbox/test", False)
    test_dl  = DataLoader(test_ds,BATCH_SIZE,shuffle=False)

    ae = CNN1dAE().to(DEVICE)
    print(ae)                       # モデル概要
    train(ae, train_dl, val_dl)

    # ドメイン別統計
    dom_subset=lambda d: torch.utils.data.Subset(train_ds,
                   [i for i,(_,dm) in enumerate(train_ds) if dm==d])
    src_dl = DataLoader(dom_subset("source"), BATCH_SIZE)
    tgt_dl = DataLoader(dom_subset("target"), BATCH_SIZE)
    mu_s, inv_s = get_stats(ae, src_dl)
    mu_t, inv_t = get_stats(ae, tgt_dl)

    evaluate(ae, test_dl, mu_s, inv_s, mu_t, inv_t)
```

```python
# ------------------------------------------------------------------ #
# 1. 依存ライブラリ
# ------------------------------------------------------------------ #
import os, random, time, warnings
import numpy as np
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset

warnings.filterwarnings("ignore")

# ------------------------------------------------------------------ #
# 2. ハイパーパラメータ
# ------------------------------------------------------------------ #
SR          = 16_000
FRAME_LEN   = 1024
HOP_LEN     = 512
N_MELS      = 128
N_FRAME     = 5
SEG_DIM     = N_MELS * N_FRAME
BATCH_SIZE  = 256
EPOCHS      = 100
DEVICE      = torch.device("cuda" if torch.cuda.is_available() else
                            "mps"  if torch.backends.mps.is_available()
                                    else "cpu")

# ------------------------------------------------------------------ #
# 3. 乱数固定
# ------------------------------------------------------------------ #
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

# ------------------------------------------------------------------ #
# 4. データセット
# ------------------------------------------------------------------ #
class LogMel5FrameDataset(Dataset):
    def __init__(self, folder, is_train=True):
        self.is_train = is_train
        self.segments = []

        file_paths = sorted([os.path.join(folder, f) for f in os.listdir(folder)
                             if f.endswith(".wav")])
        for fid, path in enumerate(file_paths):
            y, _ = librosa.load(path, sr=SR)
            mel = librosa.feature.melspectrogram(
                y=y, sr=SR, n_fft=FRAME_LEN, hop_length=HOP_LEN,
                n_mels=N_MELS, power=2.0
            )
            mel = np.log10(mel + 1e-12)
            mel = (mel - mel.mean()) / (mel.std() + 1e-9)

            T = mel.shape[1] - (mel.shape[1] % N_FRAME)
            mel = mel[:, :T].reshape(N_MELS, -1, N_FRAME)
            mel = mel.transpose(1, 0, 2)  # (num_seg, 128, 5)

            domain = "source" if "_source_" in path else "target"
            label  = 1 if (not is_train and "_anomaly" in path) else 0

            for vec in mel:
                entry = dict(vec=vec.astype(np.float32),
                             domain=domain,
                             label=label,
                             fid=fid)
                self.segments.append(entry)

    def __len__(self):
        return len(self.segments)

    def __getitem__(self, idx):
        s = self.segments[idx]
        v = torch.from_numpy(s["vec"]).unsqueeze(0)  # (1, 128, 5)
        if self.is_train:
            return v, s["domain"]
        else:
            return v, s["label"], s["domain"], s["fid"]

# ------------------------------------------------------------------ #
# 5. LSTMオートエンコーダ
# ------------------------------------------------------------------ #
class LSTMAE(nn.Module):
    def __init__(self, input_dim=128, hidden_dim=64, bottleneck_dim=8, num_layers=2):
        super().__init__()
        self.encoder_lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc_enc = nn.Linear(hidden_dim, bottleneck_dim)

        self.fc_dec = nn.Linear(bottleneck_dim, hidden_dim)
        self.decoder_lstm = nn.LSTM(hidden_dim, input_dim, num_layers, batch_first=True)

    def forward(self, x):
        x = x.squeeze(1).permute(0, 2, 1)  # (B, 5, 128)

        _, (h_n, _) = self.encoder_lstm(x)
        h = h_n[-1]  # (B, hidden_dim)
        z = self.fc_enc(h)  # (B, bottleneck_dim)

        h_dec = self.fc_dec(z).unsqueeze(1).repeat(1, x.size(1), 1)  # (B, 5, hidden_dim)
        out, _ = self.decoder_lstm(h_dec)  # (B, 5, 128)

        out = out.permute(0, 2, 1).unsqueeze(1)  # (B, 1, 128, 5)
        return out

# ------------------------------------------------------------------ #
# 6. 学習
# ------------------------------------------------------------------ #
def train(model, train_loader, val_loader):
    crit = nn.MSELoss()
    opt  = optim.Adam(model.parameters(), lr=1e-3)
    tr_hist, vl_hist = [], []

    for epoch in range(1, EPOCHS+1):
        model.train()
        tr_loss = 0.0
        for vec, _ in train_loader:
            vec = vec.to(DEVICE)
            loss = crit(model(vec), vec)
            opt.zero_grad(); loss.backward(); opt.step()
            tr_loss += loss.item()
        tr_loss /= len(train_loader)

        model.eval()
        vl_loss = 0.0
        with torch.no_grad():
            for vec, _ in val_loader:
                vec = vec.to(DEVICE)
                vl_loss += crit(model(vec), vec).item()
        vl_loss /= len(val_loader)

        tr_hist.append(tr_loss); vl_hist.append(vl_loss)
        print(f"Epoch {epoch:3d}/{EPOCHS}  Train {tr_loss:.4f}  Val {vl_loss:.4f}")
    return tr_hist, vl_hist

# ------------------------------------------------------------------ #
# 7. マハラノビス
# ------------------------------------------------------------------ #
def stats_by_domain(model, loader):
    model.eval(); errs = []
    with torch.no_grad():
        for vec, _ in loader:
            vec = vec.to(DEVICE)
            diff = (vec - model(vec)).cpu().numpy()
            diff = diff.reshape(diff.shape[0], -1)
            errs.append(diff)
    errs = np.concatenate(errs, 0)
    mu  = errs.mean(0, keepdims=True)
    cov = np.cov(errs.T) + 1e-6*np.eye(errs.shape[1])
    inv = np.linalg.inv(cov)
    return mu, inv

def maha(vec, mu, inv):
    d = vec - mu
    return float(d @ inv @ d.T)

# ------------------------------------------------------------------ #
# 8. 評価
# ------------------------------------------------------------------ #
def evaluate(model, loader, mu_s, inv_s, mu_t, inv_t):
    model.eval()
    seg_scores, seg_labels, seg_fids = [], [], []

    with torch.no_grad():
        for vec, label, domain, fid in tqdm(loader, desc="Eval"):
            vec = vec.to(DEVICE)
            diff = (vec - model(vec)).cpu().numpy()
            diff = diff.reshape(diff.shape[0], -1)
            for v, f in zip(diff, fid.numpy()):
                Ds = maha(v, mu_s, inv_s)
                Dt = maha(v, mu_t, inv_t)
                seg_scores.append(min(Ds, Dt))
            seg_labels.extend(label.numpy())
            seg_fids.extend(fid.numpy())

    file_scores, file_labels = {}, {}
    for sc, lb, fid in zip(seg_scores, seg_labels, seg_fids):
        file_scores.setdefault(fid, []).append(sc)
        file_labels[fid] = lb
    scores = np.array([np.mean(v) for v in file_scores.values()])
    labels = np.array([file_labels[k] for k in sorted(file_scores)])

    auc = roc_auc_score(labels, scores)
    print(f"\nFile‑level AUC = {auc:.4f}")

    plt.figure(figsize=(7,4))
    plt.hist(scores[labels==0], 50, alpha=.6, label="Normal", density=True)
    plt.hist(scores[labels==1], 50, alpha=.6, label="Anomaly", density=True)
    plt.title("Mahalanobis‑Distance Histogram")
    plt.xlabel("Distance")
    plt.legend(); plt.grid(); plt.tight_layout(); plt.show()

# ------------------------------------------------------------------ #
# 9. メイン
# ------------------------------------------------------------------ #
if __name__ == "__main__":
    print("device =", DEVICE)
    set_seed()

    train_folders = ["dev/gearbox/train"]
    train_dss = [LogMel5FrameDataset(f, True) for f in train_folders]
    train_full = ConcatDataset(train_dss)

    tr_size = int(0.8 * len(train_full))
    va_size = len(train_full) - tr_size
    train_ds, val_ds = random_split(train_full, [tr_size, va_size])

    train_loader = DataLoader(train_ds, BATCH_SIZE, shuffle=True, drop_last=True)
    val_loader   = DataLoader(val_ds, BATCH_SIZE, shuffle=False)
    test_ds = LogMel5FrameDataset("dev/gearbox/test", is_train=False)
    test_loader = DataLoader(test_ds, BATCH_SIZE, shuffle=False)

    ae = LSTMAE().to(DEVICE)
    train_hist, val_hist = train(ae, train_loader, val_loader)

    def domain_subset(dataset, dom):
        idx = [i for i, (_, d) in enumerate(dataset) if d == dom]
        return torch.utils.data.Subset(dataset, idx)

    src_loader = DataLoader(domain_subset(train_ds, "source"), BATCH_SIZE, shuffle=False)
    tgt_loader = DataLoader(domain_subset(train_ds, "target"), BATCH_SIZE, shuffle=False)

    mu_s, inv_s = stats_by_domain(ae, src_loader)
    mu_t, inv_t = stats_by_domain(ae, tgt_loader)

    evaluate(ae, test_loader, mu_s, inv_s, mu_t, inv_t)
```

```python
"""
DCASE‑style First‑Shot Autoencoder (MFCC版)
==========================================
Harada et al., EUSIPCO 2023 を参考にした全結合 Autoencoder による異常検知。
Log-Mel ではなく MFCC (e.g., 20次元) を使用。

・前処理  : STFT 64 ms / 50% Hop → MFCC → 5 フレーム連結
・モデル  : Linear‑AE (5×20=100次元 → 128 → ... → 8 → ... → 100)
・スコア  :  min{D_s , D_t}（Mahalanobis 距離）
・AUC     : ファイル単位（セグメント平均）
"""

# ------------------------------------------------------------------ #
# 1. 依存ライブラリ
# ------------------------------------------------------------------ #
import os, random, time, warnings
import numpy as np
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset

warnings.filterwarnings("ignore")

# ------------------------------------------------------------------ #
# 2. ハイパーパラメータ
# ------------------------------------------------------------------ #
SR          = 16_000
FRAME_LEN   = 1024            # 64 ms
HOP_LEN     = 512             # 50% hop
N_MFCC      = 20              # ← MFCC 次元数
N_FRAME     = 5
SEG_DIM     = N_MFCC * N_FRAME

BATCH_SIZE  = 256
EPOCHS      = 100
DEVICE      = torch.device("cuda" if torch.cuda.is_available() else
                            "mps"  if torch.backends.mps.is_available()
                                    else "cpu")

# ------------------------------------------------------------------ #
# 3. 乱数固定
# ------------------------------------------------------------------ #
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

# ------------------------------------------------------------------ #
# 4. データセット（MFCC版）
# ------------------------------------------------------------------ #
class MFCC5FrameDataset(Dataset):
    def __init__(self, folder, is_train=True):
        self.is_train = is_train
        self.segments = []

        file_paths = sorted([os.path.join(folder, f) for f in os.listdir(folder)
                             if f.endswith(".wav")])
        for fid, path in enumerate(file_paths):
            y, _ = librosa.load(path, sr=SR)
            mfcc = librosa.feature.mfcc(
                y=y, sr=SR, n_mfcc=N_MFCC, n_fft=FRAME_LEN,
                hop_length=HOP_LEN
            )
            mfcc = (mfcc - mfcc.mean()) / (mfcc.std() + 1e-9)

            T = mfcc.shape[1] - (mfcc.shape[1] % N_FRAME)
            mfcc = mfcc[:, :T].reshape(N_MFCC, -1, N_FRAME)
            mfcc = mfcc.transpose(1, 0, 2).reshape(-1, SEG_DIM)

            domain = "source" if "_source_" in path else "target"
            label  = 1 if (not is_train and "_anomaly" in path) else 0

            for vec in mfcc:
                entry = dict(vec=vec.astype(np.float32),
                             domain=domain,
                             label=label,
                             fid=fid)
                self.segments.append(entry)

    def __len__(self):
        return len(self.segments)

    def __getitem__(self, idx):
        s = self.segments[idx]
        v = torch.from_numpy(s["vec"])
        if self.is_train:
            return v, s["domain"]
        else:
            return v, s["label"], s["domain"], s["fid"]

# ------------------------------------------------------------------ #
# 5. 全結合オートエンコーダ
# ------------------------------------------------------------------ #
class LinearAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(SEG_DIM, 128), nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, 128),     nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, 128),     nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128,   8)
        )
        self.decoder = nn.Sequential(
            nn.Linear( 8, 128),      nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, 128),     nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, 128),     nn.BatchNorm1d(128), nn.ReLU(),
            nn.Linear(128, SEG_DIM)
        )

    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z)

# ------------------------------------------------------------------ #
# 6. 学習
# ------------------------------------------------------------------ #
def train(model, train_loader, val_loader):
    crit = nn.MSELoss()
    opt  = optim.Adam(model.parameters(), lr=1e-3)
    tr_hist, vl_hist = [], []

    for epoch in range(1, EPOCHS+1):
        model.train()
        tr_loss = 0.0
        for vec, _ in train_loader:
            vec = vec.to(DEVICE)
            loss = crit(model(vec), vec)
            opt.zero_grad(); loss.backward(); opt.step()
            tr_loss += loss.item()
        tr_loss /= len(train_loader)

        model.eval()
        vl_loss = 0.0
        with torch.no_grad():
            for vec, _ in val_loader:
                vec = vec.to(DEVICE)
                vl_loss += crit(model(vec), vec).item()
        vl_loss /= len(val_loader)

        tr_hist.append(tr_loss); vl_hist.append(vl_loss)
        print(f"Epoch {epoch:3d}/{EPOCHS}  Train {tr_loss:.4f}  Val {vl_loss:.4f}")
    return tr_hist, vl_hist

# ------------------------------------------------------------------ #
# 7. マハラノビス距離
# ------------------------------------------------------------------ #
def stats_by_domain(model, loader):
    model.eval(); errs = []
    with torch.no_grad():
        for vec, _ in loader:
            vec = vec.to(DEVICE)
            diff = (vec - model(vec)).cpu().numpy()
            errs.append(diff)
    errs = np.concatenate(errs, 0)
    mu  = errs.mean(0, keepdims=True)
    cov = np.cov(errs.T) + 1e-6*np.eye(SEG_DIM)
    inv = np.linalg.inv(cov)
    return mu, inv

def maha(vec, mu, inv):
    d = vec - mu
    return float(d @ inv @ d.T)

# ------------------------------------------------------------------ #
# 8. 評価
# ------------------------------------------------------------------ #
def evaluate(model, loader, mu_s, inv_s, mu_t, inv_t):
    model.eval()
    seg_scores, seg_labels, seg_fids = [], [], []

    with torch.no_grad():
        for vec, label, domain, fid in tqdm(loader, desc="Eval"):
            vec = vec.to(DEVICE)
            diff = (vec - model(vec)).cpu().numpy()
            for v, f in zip(diff, fid.numpy()):
                Ds = maha(v, mu_s, inv_s)
                Dt = maha(v, mu_t, inv_t)
                seg_scores.append(min(Ds, Dt))
            seg_labels.extend(label.numpy())
            seg_fids.extend(fid.numpy())

    file_scores = {}
    file_labels = {}
    for sc, lb, fid in zip(seg_scores, seg_labels, seg_fids):
        file_scores.setdefault(fid, []).append(sc)
        file_labels[fid] = lb
    scores = np.array([np.mean(v) for v in file_scores.values()])
    labels = np.array([file_labels[k] for k in sorted(file_scores)])

    auc = roc_auc_score(labels, scores)
    print(f"\nFile-level AUC = {auc:.4f}")

    plt.figure(figsize=(7,4))
    plt.hist(scores[labels==0], 50, alpha=.6, label="Normal", density=True)
    plt.hist(scores[labels==1], 50, alpha=.6, label="Anomaly",density=True)
    plt.title("Mahalanobis Distance Histogram"); plt.xlabel("Distance")
    plt.legend(); plt.grid(); plt.tight_layout(); plt.show()

# ------------------------------------------------------------------ #
# 9. メイン処理
# ------------------------------------------------------------------ #
if __name__ == "__main__":
    print("device =", DEVICE)
    set_seed()

    train_folders = ["dev/gearbox/train"]  # ← 必要に応じてパス変更
    train_dss = [MFCC5FrameDataset(f, True) for f in train_folders]
    train_full = ConcatDataset(train_dss)

    tr_size = int(0.8 * len(train_full))
    va_size = len(train_full) - tr_size
    train_ds, val_ds = random_split(train_full, [tr_size, va_size])

    train_loader = DataLoader(train_ds, BATCH_SIZE, shuffle=True, drop_last=True)
    val_loader   = DataLoader(val_ds, BATCH_SIZE, shuffle=False)

    test_ds = MFCC5FrameDataset("dev/gearbox/test", is_train=False)
    test_loader = DataLoader(test_ds, BATCH_SIZE, shuffle=False)

    ae = LinearAE().to(DEVICE)
    train_hist, val_hist = train(ae, train_loader, val_loader)

    def domain_subset(dataset, dom):
        idx = [i for i, (_, d) in enumerate(dataset) if d == dom]
        return torch.utils.data.Subset(dataset, idx)

    src_loader = DataLoader(domain_subset(train_ds, "source"), BATCH_SIZE, shuffle=False)
    tgt_loader = DataLoader(domain_subset(train_ds, "target"), BATCH_SIZE, shuffle=False)

    mu_s, inv_s = stats_by_domain(ae, src_loader)
    mu_t, inv_t = stats_by_domain(ae, tgt_loader)

    evaluate(ae, test_loader, mu_s, inv_s, mu_t, inv_t)
```

```python
"""
Conv1d-AE for MFCC Time-Series Anomaly Detection
================================================
Harada et al., EUSIPCO 2023 風の異常検知を参考に、Conv1d を使用して
MFCC の時間的変化を捉える構成に変更。
"""

# ------------------------------------------------------------------ #
# 1. 依存ライブラリ
# ------------------------------------------------------------------ #
import os, random, time, warnings
import numpy as np
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset

warnings.filterwarnings("ignore")

# ------------------------------------------------------------------ #
# 2. ハイパーパラメータ
# ------------------------------------------------------------------ #
SR          = 16_000
FRAME_LEN   = 1024
HOP_LEN     = 512
N_MFCC      = 20

BATCH_SIZE  = 16  # 長さが可変なため小さめ
EPOCHS      = 50
DEVICE      = torch.device("cuda" if torch.cuda.is_available() else
                            "mps"  if torch.backends.mps.is_available()
                                    else "cpu")

# ------------------------------------------------------------------ #
# 3. 乱数固定
# ------------------------------------------------------------------ #
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

# ------------------------------------------------------------------ #
# 4. データセット（MFCC全時系列）
# ------------------------------------------------------------------ #
class MFCCDataset(Dataset):
    def __init__(self, folder, is_train=True):
        self.is_train = is_train
        self.data = []

        file_paths = sorted([os.path.join(folder, f) for f in os.listdir(folder)
                             if f.endswith(".wav")])
        for fid, path in enumerate(file_paths):
            y, _ = librosa.load(path, sr=SR)
            mfcc = librosa.feature.mfcc(
                y=y, sr=SR, n_mfcc=N_MFCC, n_fft=FRAME_LEN,
                hop_length=HOP_LEN
            )
            mfcc = (mfcc - mfcc.mean()) / (mfcc.std() + 1e-9)

            domain = "source" if "_source_" in path else "target"
            label  = 1 if (not is_train and "_anomaly" in path) else 0

            self.data.append(dict(
                vec=mfcc.astype(np.float32),
                domain=domain,
                label=label,
                fid=fid
            ))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        s = self.data[idx]
        v = torch.from_numpy(s["vec"])  # shape: (20, T)
        if self.is_train:
            return v, s["domain"]
        else:
            return v, s["label"], s["domain"], s["fid"]

# ------------------------------------------------------------------ #
# 5. Conv1d オートエンコーダ
# ------------------------------------------------------------------ #
class Conv1dAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv1d(N_MFCC, 32, kernel_size=5, padding=2), nn.ReLU(),
            nn.Conv1d(32, 16, kernel_size=5, padding=2), nn.ReLU(),
            nn.Conv1d(16, 8, kernel_size=5, padding=2), nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Conv1d(8, 16, kernel_size=5, padding=2), nn.ReLU(),
            nn.Conv1d(16, 32, kernel_size=5, padding=2), nn.ReLU(),
            nn.Conv1d(32, N_MFCC, kernel_size=5, padding=2)
        )

    def forward(self, x):
        return self.decoder(self.encoder(x))

# ------------------------------------------------------------------ #
# 6. 学習
# ------------------------------------------------------------------ #
def train(model, train_loader, val_loader):
    crit = nn.MSELoss()
    opt  = optim.Adam(model.parameters(), lr=1e-3)
    tr_hist, vl_hist = [], []

    for epoch in range(1, EPOCHS+1):
        model.train()
        tr_loss = 0.0
        for vec, _ in train_loader:
            vec = vec.to(DEVICE)
            rec = model(vec)
            loss = crit(rec, vec)
            opt.zero_grad(); loss.backward(); opt.step()
            tr_loss += loss.item()
        tr_loss /= len(train_loader)

        model.eval()
        vl_loss = 0.0
        with torch.no_grad():
            for vec, _ in val_loader:
                vec = vec.to(DEVICE)
                rec = model(vec)
                vl_loss += crit(rec, vec).item()
        vl_loss /= len(val_loader)

        tr_hist.append(tr_loss); vl_hist.append(vl_loss)
        print(f"Epoch {epoch:3d}/{EPOCHS}  Train {tr_loss:.4f}  Val {vl_loss:.4f}")
    return tr_hist, vl_hist

# ------------------------------------------------------------------ #
# 7. マハラノビス距離
# ------------------------------------------------------------------ #
def stats_by_domain(model, loader):
    model.eval(); errs = []
    with torch.no_grad():
        for vec, _ in loader:
            vec = vec.to(DEVICE)
            rec = model(vec)
            diff = (vec - rec).cpu().numpy()  # (B, 20, T)
            B, C, T = diff.shape
            diff = diff.transpose(0, 2, 1).reshape(-1, C)  # (B*T, 20)
            errs.append(diff)
    errs = np.concatenate(errs, 0)
    mu  = errs.mean(0, keepdims=True)
    cov = np.cov(errs.T) + 1e-6*np.eye(N_MFCC)
    inv = np.linalg.inv(cov)
    return mu, inv

def maha(vec, mu, inv):
    d = vec - mu
    return float(d @ inv @ d.T)

# ------------------------------------------------------------------ #
# 8. 評価
# ------------------------------------------------------------------ #
def evaluate(model, loader, mu_s, inv_s, mu_t, inv_t):
    model.eval()
    file_scores = {}
    file_labels = {}

    with torch.no_grad():
        for vec, label, domain, fid in tqdm(loader, desc="Eval"):
            vec = vec.to(DEVICE)
            rec = model(vec)
            diff = (vec - rec).cpu().numpy()  # (B, 20, T)
            B, C, T = diff.shape
            diff = diff.transpose(0, 2, 1).reshape(B * T, C)

            for b in range(B):
                start = b * T
                end = start + T
                scores = [min(maha(d, mu_s, inv_s), maha(d, mu_t, inv_t)) for d in diff[start:end]]
                score = np.mean(scores)
                file_scores[fid[b].item()] = score
                file_labels[fid[b].item()] = label[b].item()

    scores = np.array([file_scores[k] for k in sorted(file_scores)])
    labels = np.array([file_labels[k] for k in sorted(file_labels)])

    auc = roc_auc_score(labels, scores)
    print(f"\nFile-level AUC = {auc:.4f}")

    plt.figure(figsize=(7,4))
    plt.hist(scores[labels==0], 50, alpha=.6, label="Normal", density=True)
    plt.hist(scores[labels==1], 50, alpha=.6, label="Anomaly", density=True)
    plt.title("Mahalanobis Distance Histogram"); plt.xlabel("Distance")
    plt.legend(); plt.grid(); plt.tight_layout(); plt.show()

# ------------------------------------------------------------------ #
# 9. メイン処理
# ------------------------------------------------------------------ #
if __name__ == "__main__":
    print("device =", DEVICE)
    set_seed()

    train_folders = ["dev/gearbox/train"]
    train_dss = [MFCCDataset(f, True) for f in train_folders]
    train_full = ConcatDataset(train_dss)

    tr_size = int(0.8 * len(train_full))
    va_size = len(train_full) - tr_size
    train_ds, val_ds = random_split(train_full, [tr_size, va_size])

    train_loader = DataLoader(train_ds, BATCH_SIZE, shuffle=True, drop_last=True)
    val_loader   = DataLoader(val_ds, BATCH_SIZE, shuffle=False)

    test_ds = MFCCDataset("dev/gearbox/test", is_train=False)
    test_loader = DataLoader(test_ds, BATCH_SIZE, shuffle=False)

    ae = Conv1dAE().to(DEVICE)
    train_hist, val_hist = train(ae, train_loader, val_loader)

    def domain_subset(dataset, dom):
        idx = [i for i, (x, d) in enumerate(dataset) if d == dom]
        return torch.utils.data.Subset(dataset, idx)

    src_loader = DataLoader(domain_subset(train_ds, "source"), BATCH_SIZE, shuffle=False)
    tgt_loader = DataLoader(domain_subset(train_ds, "target"), BATCH_SIZE, shuffle=False)

    mu_s, inv_s = stats_by_domain(ae, src_loader)
    mu_t, inv_t = stats_by_domain(ae, tgt_loader)

    evaluate(ae, test_loader, mu_s, inv_s, mu_t, inv_t)
```

```python
"""
Conv1d-AE for MFCC Time-Series Anomaly Detection
================================================
Harada et al., EUSIPCO 2023 風の異常検知を参考に、Conv1d を使用して
MFCC の時間的変化を捉える構成に変更。
"""

# ------------------------------------------------------------------ #
# 1. 依存ライブラリ
# ------------------------------------------------------------------ #
import os, random, time, warnings
import numpy as np
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset

warnings.filterwarnings("ignore")

# ------------------------------------------------------------------ #
# 2. ハイパーパラメータ
# ------------------------------------------------------------------ #
SR          = 16_000
FRAME_LEN   = 1024
HOP_LEN     = 512
N_MFCC      = 16

BATCH_SIZE  = 16  # 長さが可変なため小さめ
EPOCHS      = 100
DEVICE      = torch.device("cuda" if torch.cuda.is_available() else
                            "mps"  if torch.backends.mps.is_available()
                                    else "cpu")

# ------------------------------------------------------------------ #
# 3. 乱数固定
# ------------------------------------------------------------------ #
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

# ------------------------------------------------------------------ #
# 4. データセット（MFCC全時系列）
# ------------------------------------------------------------------ #
class MFCCDataset(Dataset):
    def __init__(self, folder, is_train=True):
        self.is_train = is_train
        self.data = []

        file_paths = sorted([os.path.join(folder, f) for f in os.listdir(folder)
                             if f.endswith(".wav")])
        for fid, path in enumerate(file_paths):
            y, _ = librosa.load(path, sr=SR)
            mfcc = librosa.feature.mfcc(
                y=y, sr=SR, n_mfcc=N_MFCC, n_fft=FRAME_LEN,
                hop_length=HOP_LEN
            )
            mfcc = (mfcc - mfcc.mean()) / (mfcc.std() + 1e-9)

            domain = "source" if "_source_" in path else "target"
            label  = 1 if (not is_train and "_anomaly" in path) else 0

            self.data.append(dict(
                vec=mfcc.astype(np.float32),
                domain=domain,
                label=label,
                fid=fid
            ))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        s = self.data[idx]
        v = torch.from_numpy(s["vec"])  # shape: (20, T)
        if self.is_train:
            return v, s["domain"]
        else:
            return v, s["label"], s["domain"], s["fid"]

# ------------------------------------------------------------------ #
# 5. Conv1d オートエンコーダ
# ------------------------------------------------------------------ #
class Conv1dAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv1d(N_MFCC, 32, kernel_size=5, padding=2), nn.ReLU(),
            nn.Conv1d(32, 16, kernel_size=5, padding=2), nn.ReLU(),
            nn.Conv1d(16, 8, kernel_size=5, padding=2), nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Conv1d(8, 16, kernel_size=5, padding=2), nn.ReLU(),
            nn.Conv1d(16, 32, kernel_size=5, padding=2), nn.ReLU(),
            nn.Conv1d(32, N_MFCC, kernel_size=5, padding=2)
        )

    def forward(self, x):
        return self.decoder(self.encoder(x))

# ------------------------------------------------------------------ #
# 6. 学習
# ------------------------------------------------------------------ #
def train(model, train_loader, val_loader):
    crit = nn.MSELoss()
    opt  = optim.Adam(model.parameters(), lr=1e-3)
    tr_hist, vl_hist = [], []

    for epoch in range(1, EPOCHS+1):
        model.train()
        tr_loss = 0.0
        for vec, _ in train_loader:
            vec = vec.to(DEVICE)
            rec = model(vec)
            loss = crit(rec, vec)
            opt.zero_grad(); loss.backward(); opt.step()
            tr_loss += loss.item()
        tr_loss /= len(train_loader)

        model.eval()
        vl_loss = 0.0
        with torch.no_grad():
            for vec, _ in val_loader:
                vec = vec.to(DEVICE)
                rec = model(vec)
                vl_loss += crit(rec, vec).item()
        vl_loss /= len(val_loader)

        tr_hist.append(tr_loss); vl_hist.append(vl_loss)
        print(f"Epoch {epoch:3d}/{EPOCHS}  Train {tr_loss:.4f}  Val {vl_loss:.4f}")
    return tr_hist, vl_hist

# ------------------------------------------------------------------ #
# 7. マハラノビス距離
# ------------------------------------------------------------------ #
def stats_by_domain(model, loader):
    model.eval(); errs = []
    with torch.no_grad():
        for vec, _ in loader:
            vec = vec.to(DEVICE)
            rec = model(vec)
            diff = (vec - rec).cpu().numpy()  # (B, 20, T)
            B, C, T = diff.shape
            diff = diff.transpose(0, 2, 1).reshape(-1, C)  # (B*T, 20)
            errs.append(diff)
    errs = np.concatenate(errs, 0)
    mu  = errs.mean(0, keepdims=True)
    cov = np.cov(errs.T) + 1e-6*np.eye(N_MFCC)
    inv = np.linalg.inv(cov)
    return mu, inv

def maha(vec, mu, inv):
    d = vec - mu
    return float(d @ inv @ d.T)

# ------------------------------------------------------------------ #
# 8. 評価
# ------------------------------------------------------------------ #
def evaluate(model, loader, mu_s, inv_s, mu_t, inv_t):
    model.eval()
    file_scores = {}
    file_labels = {}

    with torch.no_grad():
        for vec, label, domain, fid in tqdm(loader, desc="Eval"):
            vec = vec.to(DEVICE)
            rec = model(vec)
            diff = (vec - rec).cpu().numpy()  # (B, 20, T)
            B, C, T = diff.shape
            diff = diff.transpose(0, 2, 1).reshape(B * T, C)

            for b in range(B):
                start = b * T
                end = start + T
                scores = [min(maha(d, mu_s, inv_s), maha(d, mu_t, inv_t)) for d in diff[start:end]]
                score = np.mean(scores)
                file_scores[fid[b].item()] = score
                file_labels[fid[b].item()] = label[b].item()

    scores = np.array([file_scores[k] for k in sorted(file_scores)])
    labels = np.array([file_labels[k] for k in sorted(file_labels)])

    auc = roc_auc_score(labels, scores)
    print(f"\nFile-level AUC = {auc:.4f}")

    plt.figure(figsize=(7,4))
    plt.hist(scores[labels==0], 50, alpha=.6, label="Normal", density=True)
    plt.hist(scores[labels==1], 50, alpha=.6, label="Anomaly", density=True)
    plt.title("Mahalanobis Distance Histogram"); plt.xlabel("Distance")
    plt.legend(); plt.grid(); plt.tight_layout(); plt.show()

# ------------------------------------------------------------------ #
# 9. メイン処理
# ------------------------------------------------------------------ #
if __name__ == "__main__":
    print("device =", DEVICE)
    set_seed()

    train_folders = ["dev/gearbox/train"]
    train_dss = [MFCCDataset(f, True) for f in train_folders]
    train_full = ConcatDataset(train_dss)

    tr_size = int(0.8 * len(train_full))
    va_size = len(train_full) - tr_size
    train_ds, val_ds = random_split(train_full, [tr_size, va_size])

    train_loader = DataLoader(train_ds, BATCH_SIZE, shuffle=True, drop_last=True)
    val_loader   = DataLoader(val_ds, BATCH_SIZE, shuffle=False)

    test_ds = MFCCDataset("dev/gearbox/test", is_train=False)
    test_loader = DataLoader(test_ds, BATCH_SIZE, shuffle=False)

    ae = Conv1dAE().to(DEVICE)
    train_hist, val_hist = train(ae, train_loader, val_loader)

    def domain_subset(dataset, dom):
        idx = [i for i, (x, d) in enumerate(dataset) if d == dom]
        return torch.utils.data.Subset(dataset, idx)

    src_loader = DataLoader(domain_subset(train_ds, "source"), BATCH_SIZE, shuffle=False)
    tgt_loader = DataLoader(domain_subset(train_ds, "target"), BATCH_SIZE, shuffle=False)

    mu_s, inv_s = stats_by_domain(ae, src_loader)
    mu_t, inv_t = stats_by_domain(ae, tgt_loader)

    evaluate(ae, test_loader, mu_s, inv_s, mu_t, inv_t)
```

```python
"""
LSTM-AE for MFCC Time-Series Anomaly Detection
==============================================
Conv1d の代わりに LSTM を用いた Autoencoder 構成。
MFCC の時系列を直接 RNN で処理。
"""

# ------------------------------------------------------------------ #
# 1. 依存ライブラリ
# ------------------------------------------------------------------ #
import os, random, warnings
import numpy as np
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split, ConcatDataset

warnings.filterwarnings("ignore")

# ------------------------------------------------------------------ #
# 2. ハイパーパラメータ
# ------------------------------------------------------------------ #
SR          = 16_000
FRAME_LEN   = 1024
HOP_LEN     = 512
N_MFCC      = 16

BATCH_SIZE  = 16
EPOCHS      = 100
DEVICE      = torch.device("cuda" if torch.cuda.is_available() else
                            "mps"  if torch.backends.mps.is_available()
                                    else "cpu")

# ------------------------------------------------------------------ #
# 3. 乱数固定
# ------------------------------------------------------------------ #
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

# ------------------------------------------------------------------ #
# 4. データセット（MFCC 全時系列）
# ------------------------------------------------------------------ #
class MFCCDataset(Dataset):
    def __init__(self, folder, is_train=True):
        self.is_train = is_train
        self.data = []

        file_paths = sorted([os.path.join(folder, f) for f in os.listdir(folder)
                             if f.endswith(".wav")])
        for fid, path in enumerate(file_paths):
            y, _ = librosa.load(path, sr=SR)
            mfcc = librosa.feature.mfcc(
                y=y, sr=SR, n_mfcc=N_MFCC, n_fft=FRAME_LEN,
                hop_length=HOP_LEN
            )
            mfcc = (mfcc - mfcc.mean()) / (mfcc.std() + 1e-9)

            domain = "source" if "_source_" in path else "target"
            label  = 1 if (not is_train and "_anomaly" in path) else 0

            self.data.append(dict(
                vec=mfcc.astype(np.float32),
                domain=domain,
                label=label,
                fid=fid
            ))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        s = self.data[idx]
        v = torch.from_numpy(s["vec"])  # shape: (C, T)
        if self.is_train:
            return v, s["domain"]
        else:
            return v, s["label"], s["domain"], s["fid"]

# ------------------------------------------------------------------ #
# 5. LSTM オートエンコーダ
# ------------------------------------------------------------------ #
class LSTMAE(nn.Module):
    def __init__(self, input_dim=N_MFCC, hidden_dim=64, bottleneck_dim=16, num_layers=2):
        super().__init__()
        self.encoder = nn.LSTM(input_dim, hidden_dim, num_layers=num_layers,
                               batch_first=True, bidirectional=False)
        self.bottleneck = nn.Linear(hidden_dim, bottleneck_dim)
        self.unbottleneck = nn.Linear(bottleneck_dim, hidden_dim)
        self.decoder = nn.LSTM(hidden_dim, input_dim, num_layers=num_layers,
                               batch_first=True, bidirectional=False)

    def forward(self, x):
        # x: (B, C, T) → (B, T, C)
        x = x.transpose(1, 2)

        # Encode
        enc_out, _ = self.encoder(x)  # (B, T, H)
        z = self.bottleneck(enc_out)  # (B, T, bottleneck)
        z = self.unbottleneck(z)      # (B, T, H)

        # Decode
        dec_out, _ = self.decoder(z)  # (B, T, C)

        return dec_out.transpose(1, 2)  # → (B, C, T)

# ------------------------------------------------------------------ #
# 6. 学習
# ------------------------------------------------------------------ #
def train(model, train_loader, val_loader):
    crit = nn.MSELoss()
    opt  = optim.Adam(model.parameters(), lr=1e-3)
    tr_hist, vl_hist = [], []

    for epoch in range(1, EPOCHS+1):
        model.train()
        tr_loss = 0.0
        for vec, _ in train_loader:
            vec = vec.to(DEVICE)
            rec = model(vec)
            loss = crit(rec, vec)
            opt.zero_grad(); loss.backward(); opt.step()
            tr_loss += loss.item()
        tr_loss /= len(train_loader)

        model.eval()
        vl_loss = 0.0
        with torch.no_grad():
            for vec, _ in val_loader:
                vec = vec.to(DEVICE)
                rec = model(vec)
                vl_loss += crit(rec, vec).item()
        vl_loss /= len(val_loader)

        tr_hist.append(tr_loss); vl_hist.append(vl_loss)
        print(f"Epoch {epoch:3d}/{EPOCHS}  Train {tr_loss:.4f}  Val {vl_loss:.4f}")
    return tr_hist, vl_hist

# ------------------------------------------------------------------ #
# 7. マハラノビス距離
# ------------------------------------------------------------------ #
def stats_by_domain(model, loader):
    model.eval(); errs = []
    with torch.no_grad():
        for vec, _ in loader:
            vec = vec.to(DEVICE)
            rec = model(vec)
            diff = (vec - rec).cpu().numpy()  # (B, C, T)
            B, C, T = diff.shape
            diff = diff.transpose(0, 2, 1).reshape(-1, C)  # (B*T, C)
            errs.append(diff)
    errs = np.concatenate(errs, 0)
    mu  = errs.mean(0, keepdims=True)
    cov = np.cov(errs.T) + 1e-6*np.eye(C)
    inv = np.linalg.inv(cov)
    return mu, inv

def maha(vec, mu, inv):
    d = vec - mu
    return float(d @ inv @ d.T)

# ------------------------------------------------------------------ #
# 8. 評価
# ------------------------------------------------------------------ #
def evaluate(model, loader, mu_s, inv_s, mu_t, inv_t):
    model.eval()
    file_scores = {}
    file_labels = {}

    with torch.no_grad():
        for vec, label, domain, fid in tqdm(loader, desc="Eval"):
            vec = vec.to(DEVICE)
            rec = model(vec)
            diff = (vec - rec).cpu().numpy()  # (B, C, T)
            B, C, T = diff.shape
            diff = diff.transpose(0, 2, 1).reshape(B * T, C)

            for b in range(B):
                start = b * T
                end = start + T
                scores = [min(maha(d, mu_s, inv_s), maha(d, mu_t, inv_t)) for d in diff[start:end]]
                score = np.mean(scores)
                file_scores[fid[b].item()] = score
                file_labels[fid[b].item()] = label[b].item()

    scores = np.array([file_scores[k] for k in sorted(file_scores)])
    labels = np.array([file_labels[k] for k in sorted(file_labels)])

    auc = roc_auc_score(labels, scores)
    print(f"\nFile-level AUC = {auc:.4f}")

    plt.figure(figsize=(7,4))
    plt.hist(scores[labels==0], 50, alpha=.6, label="Normal", density=True)
    plt.hist(scores[labels==1], 50, alpha=.6, label="Anomaly", density=True)
    plt.title("Mahalanobis Distance Histogram"); plt.xlabel("Distance")
    plt.legend(); plt.grid(); plt.tight_layout(); plt.show()

# ------------------------------------------------------------------ #
# 9. メイン処理
# ------------------------------------------------------------------ #
if __name__ == "__main__":
    print("device =", DEVICE)
    set_seed()

    train_folders = ["dev/gearbox/train"]
    train_dss = [MFCCDataset(f, True) for f in train_folders]
    train_full = ConcatDataset(train_dss)

    tr_size = int(0.8 * len(train_full))
    va_size = len(train_full) - tr_size
    train_ds, val_ds = random_split(train_full, [tr_size, va_size])

    train_loader = DataLoader(train_ds, BATCH_SIZE, shuffle=True, drop_last=True)
    val_loader   = DataLoader(val_ds, BATCH_SIZE, shuffle=False)

    test_ds = MFCCDataset("dev/gearbox/test", is_train=False)
    test_loader = DataLoader(test_ds, BATCH_SIZE, shuffle=False)

    ae = LSTMAE(input_dim=N_MFCC).to(DEVICE)
    train_hist, val_hist = train(ae, train_loader, val_loader)

    def domain_subset(dataset, dom):
        idx = [i for i, (x, d) in enumerate(dataset) if d == dom]
        return torch.utils.data.Subset(dataset, idx)

    src_loader = DataLoader(domain_subset(train_ds, "source"), BATCH_SIZE, shuffle=False)
    tgt_loader = DataLoader(domain_subset(train_ds, "target"), BATCH_SIZE, shuffle=False)

    mu_s, inv_s = stats_by_domain(ae, src_loader)
    mu_t, inv_t = stats_by_domain(ae, tgt_loader)

    evaluate(ae, test_loader, mu_s, inv_s, mu_t, inv_t)
```
