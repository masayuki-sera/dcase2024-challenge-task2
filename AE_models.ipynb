```python
# lstm
import os
import librosa
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
import warnings
import random

warnings.filterwarnings("ignore")

# ==== シード値の固定 ====
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    # PyTorch の再現性設定
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

# ==== データセット ====
class AudioDataset(Dataset):
    def __init__(self, folder_path, is_train=True, feature_type="log_mel"):
        self.folder_path = folder_path
        self.file_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".wav")]
        self.is_train = is_train
        self.feature_type = feature_type

        if not is_train:
            self.labels = [1 if "anomaly" in f.split("_")[4] else 0 for f in os.listdir(folder_path)]

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        path = self.file_paths[idx]
        y, sr = librosa.load(path, sr=SAMPLE_RATE)

        if self.feature_type == "spectrogram":
            S = np.abs(librosa.stft(y, n_fft=N_FFT, hop_length=HOP_LENGTH)) ** 2
            feature = librosa.amplitude_to_db(S, ref=np.max)
        elif self.feature_type == "mel":
            mel = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=N_FFT,
                                                 hop_length=HOP_LENGTH, n_mels=N_MELS)
            feature = mel
        elif self.feature_type == "log_mel":
            mel = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=N_FFT,
                                                 hop_length=HOP_LENGTH, n_mels=N_MELS)
            feature = librosa.power_to_db(mel, ref=np.max)
        else:
            raise ValueError("Invalid feature_type. Choose from: 'spectrogram', 'mel', 'log_mel'.")

        # 標準化
        feature = (feature - feature.mean()) / (feature.std() + 1e-9)
        feature_tensor = torch.tensor(feature.T, dtype=torch.float32)  # (T, F)

        if self.is_train:
            return feature_tensor
        else:
            label = self.labels[idx]
            return feature_tensor, label


# ==== LSTMベースのAutoencoder ====
class LSTM_Autoencoder(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers=1):
        super(LSTM_Autoencoder, self).__init__()
        self.encoder = nn.LSTM(input_dim, hidden_dim, num_layers=num_layers, batch_first=True)
        self.decoder = nn.LSTM(hidden_dim, input_dim, num_layers=num_layers, batch_first=True)

    def forward(self, x):
        # x: (B, T, F)
        _, (h_n, _) = self.encoder(x)
        # 最終ステップの隠れ状態を各タイムステップに繰り返してDecoderへ
        decoder_input = h_n[-1].unsqueeze(1).repeat(1, x.size(1), 1)
        out, _ = self.decoder(decoder_input)
        return out  # (B, T, F)


# ==== 学習関数 ====
def train_autoencoder(model, train_loader, val_loader):
    model.train()
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    train_losses = []
    val_losses = []

    pbar = tqdm(total=EPOCHS, desc="Training Progress", unit="epoch")

    for epoch in range(EPOCHS):
        start_time = time.time()
        model.train()
        train_loss = 0
        for batch in train_loader:
            batch = batch.to(DEVICE)
            output = model(batch)
            loss = criterion(output, batch)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for val_batch in val_loader:
                val_batch = val_batch.to(DEVICE)
                val_output = model(val_batch)
                val_loss += criterion(val_output, val_batch).item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        elapsed = time.time() - start_time
        tqdm.write(f"Epoch {epoch+1}/{EPOCHS} "
                   f"Train Loss: {avg_train_loss:.4f} | "
                   f"Val Loss: {avg_val_loss:.4f} | "
                   f"Time: {elapsed:.2f}s")

        pbar.update(1)

    pbar.close()
    return train_losses, val_losses


# ==== 評価関数 ====
def evaluate(model, dataloader):
    model.eval()
    criterion = nn.MSELoss(reduction="none")

    scores = []
    labels = []

    with torch.no_grad():
        for batch, label in tqdm(dataloader, desc="Evaluating"):
            batch = batch.to(DEVICE)
            output = model(batch)
            loss = criterion(output, batch)
            score = loss.mean(dim=(1, 2)).cpu().numpy()
            scores.extend(score)
            labels.extend(label.numpy())

    auc = roc_auc_score(labels, scores)
    print(f"AUC Score: {auc:.4f}")

    # ヒストグラム
    scores = np.array(scores)
    labels = np.array(labels)
    normal_scores = scores[labels == 0]
    anomaly_scores = scores[labels == 1]

    plt.figure(figsize=(10, 5))
    plt.hist(normal_scores, bins=50, alpha=0.6, label="Normal", density=True)
    plt.hist(anomaly_scores, bins=50, alpha=0.6, label="Anomaly", density=True)
    plt.xlabel("Reconstruction Error")
    plt.ylabel("Density")
    plt.title("Reconstruction Error Histogram")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


# ==== 損失グラフ描画 ====
def plot_losses(train_losses, val_losses):
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


# ==== ハイパーパラメータ ====
SAMPLE_RATE = 16000
N_MELS = 128
HOP_LENGTH = 512
N_FFT = 1024
BATCH_SIZE = 64
EPOCHS = 30
HIDDEN_DIM = 64
DEVICE = torch.device("mps" if torch.backends.mps.is_available() else "cuda" if torch.cuda.is_available() else "cpu")


# ==== メイン処理 ====
if __name__ == "__main__":
    set_seed(42)
    feature_type = "log_mel"

    full_dataset = AudioDataset("dev/ToyTrain/train", is_train=True, feature_type=feature_type)
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, drop_last=True)

    test_dataset = AudioDataset("dev/ToyTrain/test", is_train=False, feature_type=feature_type)
    test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

    model = LSTM_Autoencoder(input_dim=N_MELS, hidden_dim=HIDDEN_DIM).to(DEVICE)

    train_losses, val_losses = train_autoencoder(model, train_loader, val_loader)
    plot_losses(train_losses, val_losses)
    evaluate(model, test_loader)
```

```python
# lstm + attention
import os
import librosa
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
import warnings
import random

warnings.filterwarnings("ignore")

# ==== シード値の固定 ====
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    # PyTorch の再現性設定
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False


# ==== データセット ====
class AudioDataset(Dataset):
    def __init__(self, folder_path, is_train=True, feature_type="log_mel"):
        self.folder_path = folder_path
        self.file_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".wav")]
        self.is_train = is_train
        self.feature_type = feature_type

        if not is_train:
            self.labels = [1 if "anomaly" in f.split("_")[4] else 0 for f in os.listdir(folder_path)]

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        path = self.file_paths[idx]
        y, sr = librosa.load(path, sr=SAMPLE_RATE)

        if self.feature_type == "spectrogram":
            S = np.abs(librosa.stft(y, n_fft=N_FFT, hop_length=HOP_LENGTH)) ** 2
            feature = librosa.amplitude_to_db(S, ref=np.max)
        elif self.feature_type == "mel":
            mel = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=N_FFT,
                                                 hop_length=HOP_LENGTH, n_mels=N_MELS)
            feature = mel
        elif self.feature_type == "log_mel":
            mel = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=N_FFT,
                                                 hop_length=HOP_LENGTH, n_mels=N_MELS)
            feature = librosa.power_to_db(mel, ref=np.max)
        else:
            raise ValueError("Invalid feature_type.")

        feature = (feature - feature.mean()) / (feature.std() + 1e-9)
        feature_tensor = torch.tensor(feature.T, dtype=torch.float32)  # (T, F)

        if self.is_train:
            return feature_tensor
        else:
            label = self.labels[idx]
            return feature_tensor, label


# ==== アテンション機構 ====
class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super(Attention, self).__init__()
        self.attn = nn.Linear(hidden_dim, hidden_dim)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, hidden_states):
        # hidden_states: (B, T, H)
        energy = self.attn(hidden_states)  # (B, T, H)
        scores = torch.bmm(energy, hidden_states.transpose(1, 2))  # (B, T, T)
        weights = self.softmax(scores)  # (B, T, T)
        context = torch.bmm(weights, hidden_states)  # (B, T, H)
        return context


# ==== Attention付き LSTM Autoencoder ====
class AttentionLSTM_Autoencoder(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers=1):
        super(AttentionLSTM_Autoencoder, self).__init__()
        self.encoder = nn.LSTM(input_dim, hidden_dim, num_layers=num_layers, batch_first=True, bidirectional=False)
        self.attention = Attention(hidden_dim)
        self.decoder = nn.LSTM(hidden_dim, input_dim, num_layers=num_layers, batch_first=True)

    def forward(self, x):
        # x: (B, T, F)
        enc_out, _ = self.encoder(x)  # enc_out: (B, T, H)
        context = self.attention(enc_out)  # (B, T, H)
        out, _ = self.decoder(context)  # (B, T, F)
        return out


# ==== 学習関数 ====
def train_autoencoder(model, train_loader, val_loader):
    model.train()
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    train_losses = []
    val_losses = []

    pbar = tqdm(total=EPOCHS, desc="Training Progress", unit="epoch")

    for epoch in range(EPOCHS):
        start_time = time.time()
        model.train()
        train_loss = 0
        for batch in train_loader:
            batch = batch.to(DEVICE)
            output = model(batch)
            loss = criterion(output, batch)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for val_batch in val_loader:
                val_batch = val_batch.to(DEVICE)
                val_output = model(val_batch)
                val_loss += criterion(val_output, val_batch).item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        elapsed = time.time() - start_time
        tqdm.write(f"Epoch {epoch+1}/{EPOCHS} "
                   f"Train Loss: {avg_train_loss:.4f} | "
                   f"Val Loss: {avg_val_loss:.4f} | "
                   f"Time: {elapsed:.2f}s")

        pbar.update(1)

    pbar.close()
    return train_losses, val_losses


# ==== 評価関数 ====
def evaluate(model, dataloader):
    model.eval()
    criterion = nn.MSELoss(reduction="none")

    scores = []
    labels = []

    with torch.no_grad():
        for batch, label in tqdm(dataloader, desc="Evaluating"):
            batch = batch.to(DEVICE)
            output = model(batch)
            loss = criterion(output, batch)
            score = loss.mean(dim=(1, 2)).cpu().numpy()
            scores.extend(score)
            labels.extend(label.numpy())

    auc = roc_auc_score(labels, scores)
    print(f"AUC Score: {auc:.4f}")

    scores = np.array(scores)
    labels = np.array(labels)
    normal_scores = scores[labels == 0]
    anomaly_scores = scores[labels == 1]

    plt.figure(figsize=(10, 5))
    plt.hist(normal_scores, bins=50, alpha=0.6, label="Normal", density=True)
    plt.hist(anomaly_scores, bins=50, alpha=0.6, label="Anomaly", density=True)
    plt.xlabel("Reconstruction Error")
    plt.ylabel("Density")
    plt.title("Reconstruction Error Histogram")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


# ==== 損失グラフ描画 ====
def plot_losses(train_losses, val_losses):
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


# ==== ハイパーパラメータ ====
SAMPLE_RATE = 16000
N_MELS = 128
HOP_LENGTH = 512
N_FFT = 1024
BATCH_SIZE = 64
EPOCHS = 30
HIDDEN_DIM = 64
DEVICE = torch.device("mps" if torch.backends.mps.is_available() else "cuda" if torch.cuda.is_available() else "cpu")


# ==== メイン処理 ====
if __name__ == "__main__":
    set_seed(42)
    feature_type = "log_mel"

    full_dataset = AudioDataset("dev/ToyTrain/train", is_train=True, feature_type=feature_type)
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, drop_last=True)

    test_dataset = AudioDataset("dev/ToyTrain/test", is_train=False, feature_type=feature_type)
    test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

    model = AttentionLSTM_Autoencoder(input_dim=N_MELS, hidden_dim=HIDDEN_DIM).to(DEVICE)

    train_losses, val_losses = train_autoencoder(model, train_loader, val_loader)
    plot_losses(train_losses, val_losses)
    evaluate(model, test_loader)
```

```python
import os
import librosa
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
import warnings
import random

warnings.filterwarnings("ignore")

# ==== シード値の固定 ====
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False


# ==== データセット ====
class AudioDataset(Dataset):
    def __init__(self, folder_path, is_train=True, feature_type="mfcc"):
        self.folder_path = folder_path
        self.file_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".wav")]
        self.is_train = is_train
        self.feature_type = feature_type

        if not is_train:
            self.labels = [1 if "anomaly" in f.split("_")[4] else 0 for f in os.listdir(folder_path)]

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        path = self.file_paths[idx]
        y, sr = librosa.load(path, sr=SAMPLE_RATE)

        if self.feature_type == "mfcc":
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=N_MFCC, n_fft=N_FFT, hop_length=HOP_LENGTH)
            feature = (mfcc - mfcc.mean()) / (mfcc.std() + 1e-9)
        else:
            raise ValueError("Invalid feature_type. Only 'mfcc' supported in this version.")

        feature_tensor = torch.tensor(feature, dtype=torch.float32).unsqueeze(0)  # (1, 13, T)

        if self.is_train:
            return feature_tensor
        else:
            label = self.labels[idx]
            return feature_tensor, label


# ==== Residual Block の定義 ====
class ResidualBlock(nn.Module):
    def __init__(self, channels):
        super(ResidualBlock, self).__init__()
        self.block = nn.Sequential(
            nn.Conv2d(channels, channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(channels),
            nn.ReLU(),
            nn.Conv2d(channels, channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(channels)
        )
        self.relu = nn.ReLU()

    def forward(self, x):
        return self.relu(x + self.block(x))


# ==== Autoencoder モデル（ResidualBlock付き）====
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()

        self.encoder = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            ResidualBlock(32),  # 追加

            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )

        self.bottleneck = nn.Sequential(
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU()
        )

        self.decoder = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(256, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),

            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(128, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),

            ResidualBlock(64),  # 追加

            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(64, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),

            nn.Conv2d(32, 1, 3, padding=1)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.bottleneck(x)
        x = self.decoder(x)
        return x


# ==== 学習関数 ====
def train_autoencoder(model, train_loader, val_loader):
    model.train()
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    train_losses = []
    val_losses = []

    pbar = tqdm(total=EPOCHS, desc="Training Progress", unit="epoch")

    for epoch in range(EPOCHS):
        start_time = time.time()
        model.train()
        train_loss = 0

        for batch in train_loader:
            batch = batch.to(DEVICE)
            output = model(batch)
            loss = criterion(output, batch)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for val_batch in val_loader:
                val_batch = val_batch.to(DEVICE)
                val_output = model(val_batch)
                val_loss += criterion(val_output, val_batch).item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        elapsed = time.time() - start_time
        tqdm.write(f"Epoch {epoch+1}/{EPOCHS} Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | Time: {elapsed:.2f}s")
        pbar.update(1)

    pbar.close()
    return train_losses, val_losses


# ==== 評価関数 ====
def evaluate(model, dataloader):
    model.eval()
    criterion = nn.MSELoss(reduction="none")

    scores = []
    labels = []

    with torch.no_grad():
        for batch, label in tqdm(dataloader, desc="Evaluating"):
            batch = batch.to(DEVICE)
            output = model(batch)
            loss = criterion(output, batch)
            score = loss.mean(dim=(1, 2, 3)).cpu().numpy()
            scores.extend(score)
            labels.extend(label.numpy())

    auc = roc_auc_score(labels, scores)
    print(f"AUC Score: {auc:.4f}")

    scores = np.array(scores)
    labels = np.array(labels)
    normal_scores = scores[labels == 0]
    anomaly_scores = scores[labels == 1]

    plt.figure(figsize=(10, 5))
    plt.hist(normal_scores, bins=50, alpha=0.6, label="Normal", density=True)
    plt.hist(anomaly_scores, bins=50, alpha=0.6, label="Anomaly", density=True)
    plt.xlabel("Reconstruction Error")
    plt.ylabel("Density")
    plt.title("Reconstruction Error Histogram")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


# ==== 損失グラフ描画 ====
def plot_losses(train_losses, val_losses):
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


# ==== ハイパーパラメータ ====
SAMPLE_RATE = 16000
N_MELS = 128
N_MFCC = 16
HOP_LENGTH = 512
N_FFT = 1024
BATCH_SIZE = 64
EPOCHS = 50
DEVICE = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

# ==== メイン処理 ====
if __name__ == "__main__":
    set_seed(42)
    feature_type = "mfcc"

    full_dataset = AudioDataset("dev/ToyTrain/train", is_train=True, feature_type=feature_type)
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

    test_dataset = AudioDataset("dev/ToyTrain/test", is_train=False, feature_type=feature_type)
    test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

    model = Autoencoder().to(DEVICE)

    train_losses, val_losses = train_autoencoder(model, train_loader, val_loader)
    plot_losses(train_losses, val_losses)
    evaluate(model, test_loader)
```

```python
import os
import librosa
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
import warnings
import random
import torch.nn.functional as F

warnings.filterwarnings("ignore")

# ==== シード値の固定 ====
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False


# ==== データセット ====
class AudioDataset(Dataset):
    def __init__(self, folder_path, is_train=True, feature_type="mfcc"):
        self.folder_path = folder_path
        self.file_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".wav")]
        self.is_train = is_train
        self.feature_type = feature_type

        if not is_train:
            self.labels = [1 if "anomaly" in f.split("_")[4] else 0 for f in os.listdir(folder_path)]

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        path = self.file_paths[idx]
        y, sr = librosa.load(path, sr=SAMPLE_RATE)

        if self.feature_type == "mfcc":
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=N_MFCC, n_fft=N_FFT, hop_length=HOP_LENGTH)
            feature = (mfcc - mfcc.mean()) / (mfcc.std() + 1e-9)
        else:
            raise ValueError("Invalid feature_type. Only 'mfcc' supported in this version.")

        feature_tensor = torch.tensor(feature, dtype=torch.float32).unsqueeze(0)  # (1, 13, T)

        if self.is_train:
            return feature_tensor
        else:
            label = self.labels[idx]
            return feature_tensor, label


# ==== Residual Block の定義 ====
class ResidualBlock(nn.Module):
    def __init__(self, channels):
        super(ResidualBlock, self).__init__()
        self.block = nn.Sequential(
            nn.Conv2d(channels, channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(channels),
            nn.ReLU(),
            nn.Conv2d(channels, channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(channels)
        )
        self.relu = nn.ReLU()

    def forward(self, x):
        return self.relu(x + self.block(x))


# ==== Autoencoder（層を深くした構造）====
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()

        self.encoder = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),     # → (1, H, W) → (32, H, W)
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),                   # → (32, H/2, W/2)

            ResidualBlock(32),

            nn.Conv2d(32, 64, 3, padding=1),    # → (64, H/2, W/2)
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),                   # → (64, H/4, W/4)

            ResidualBlock(64),

            nn.Conv2d(64, 128, 3, padding=1),   # → (128, H/4, W/4)
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),                   # → (128, H/8, W/8)

            nn.Conv2d(128, 256, 3, padding=1),  # ★ 追加で深くする
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2),                   # → (256, H/16, W/16)
        )

        self.bottleneck = nn.Sequential(
            nn.Conv2d(256, 512, 3, padding=1),  # より圧縮＋高次元
            nn.BatchNorm2d(512),
            nn.ReLU()
        )

        self.decoder = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='nearest'),   # → (512, H/8, W/8)
            nn.Conv2d(512, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),

            nn.Upsample(scale_factor=2, mode='nearest'),   # → (256, H/4, W/4)
            nn.Conv2d(256, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),

            nn.Upsample(scale_factor=2, mode='nearest'),   # → (128, H/2, W/2)
            nn.Conv2d(128, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),

            ResidualBlock(64),

            nn.Upsample(scale_factor=2, mode='nearest'),   # → (64, H, W)
            nn.Conv2d(64, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),

            nn.Conv2d(32, 1, 3, padding=1)  # → (1, H, W)
        )

    def forward(self, x):
        original_size = x.size()  # (B, C, H, W)
        x = self.encoder(x)
        x = self.bottleneck(x)
        x = self.decoder(x)

        # 入力と全く同じサイズに補間（バッチ数・チャネル数は変えない）
        x = F.interpolate(x, size=(original_size[2], original_size[3]), mode='nearest')
        return x


# ==== 学習関数 ====
def train_autoencoder(model, train_loader, val_loader):
    model.train()
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    train_losses = []
    val_losses = []

    pbar = tqdm(total=EPOCHS, desc="Training Progress", unit="epoch")

    for epoch in range(EPOCHS):
        start_time = time.time()
        model.train()
        train_loss = 0

        for batch in train_loader:
            batch = batch.to(DEVICE)
            output = model(batch)
            loss = criterion(output, batch)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for val_batch in val_loader:
                val_batch = val_batch.to(DEVICE)
                val_output = model(val_batch)
                val_loss += criterion(val_output, val_batch).item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        elapsed = time.time() - start_time
        tqdm.write(f"Epoch {epoch+1}/{EPOCHS} Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | Time: {elapsed:.2f}s")
        pbar.update(1)

    pbar.close()
    return train_losses, val_losses


# ==== 評価関数 ====
def evaluate(model, dataloader):
    model.eval()
    criterion = nn.MSELoss(reduction="none")

    scores = []
    labels = []

    with torch.no_grad():
        for batch, label in tqdm(dataloader, desc="Evaluating"):
            batch = batch.to(DEVICE)
            output = model(batch)
            loss = criterion(output, batch)
            score = loss.mean(dim=(1, 2, 3)).cpu().numpy()
            scores.extend(score)
            labels.extend(label.numpy())

    auc = roc_auc_score(labels, scores)
    print(f"AUC Score: {auc:.4f}")

    scores = np.array(scores)
    labels = np.array(labels)
    normal_scores = scores[labels == 0]
    anomaly_scores = scores[labels == 1]

    plt.figure(figsize=(10, 5))
    plt.hist(normal_scores, bins=50, alpha=0.6, label="Normal", density=True)
    plt.hist(anomaly_scores, bins=50, alpha=0.6, label="Anomaly", density=True)
    plt.xlabel("Reconstruction Error")
    plt.ylabel("Density")
    plt.title("Reconstruction Error Histogram")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


# ==== 損失グラフ描画 ====
def plot_losses(train_losses, val_losses):
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


# ==== ハイパーパラメータ ====
SAMPLE_RATE = 16000
N_MELS = 128
N_MFCC = 16
HOP_LENGTH = 512
N_FFT = 1024
BATCH_SIZE = 64
EPOCHS = 50
DEVICE = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

# ==== メイン処理 ====
if __name__ == "__main__":
    set_seed(42)
    feature_type = "mfcc"

    full_dataset = AudioDataset("dev/ToyTrain/train", is_train=True, feature_type=feature_type)
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

    test_dataset = AudioDataset("dev/ToyTrain/test", is_train=False, feature_type=feature_type)
    test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

    model = Autoencoder().to(DEVICE)

    train_losses, val_losses = train_autoencoder(model, train_loader, val_loader)
    plot_losses(train_losses, val_losses)
    evaluate(model, test_loader)
```
